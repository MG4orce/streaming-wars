```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline


df = pd.read_csv(r'/Users/myron/Desktop/ds_project/dataset/MoviesOnStreamingPlatforms_updated.csv')

## Read Headers
#df[['Netflix', 'Hulu', 'Prime Video', 'Disney+']]

# Sum up the columns as a list
num_movies = [df['Netflix'].sum(), df['Hulu'].sum(), df['Prime Video'].sum(), df['Disney+'].sum()]

fig, ax = plt.subplots()    # Create figure object and an axes object
# Plotting on axes - bar method creates chart. x=where on x axis do we center the bar range(1, 5) returns a list [1, 2, 3, 4]
# height=how tall given bar has to be. num_movies is how many movies are on a platform
ax.bar(x=range(1,5), height=num_movies)
ax.set_xticks(range(1, 5))
ax.set_xticklabels(['Netflix', 'Hulu', 'Prime Video', 'Disney+'])
plt.show()

#for column, row in df.iteritems():
    #print(column[0], row)
# Read each column
#print(df[['Title', 'Country', 'Netflix']])
```

Here we can see that Amazon Prime Video holds more movies than Netflix, Hulu, and Disney+ combined.
